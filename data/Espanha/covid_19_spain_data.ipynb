```python
import pandas as pd
```

Dados retirados de https://github.com/datadista/datasets/tree/master/COVID%2019. As fontes utilizadas são o Ministerio de Sanidad, Consumo y Bienestar Social, Departamento de Seguridad Nacional, Instituto de Salud Carlos III e do Ministerio de Transportes  Movilidad y Agenda Urbana da Espanha. A base de dados contém informações sobre o número de casos e mortes diários por região da Espanha e é atualizada diariamente.

O código a seguir realiza a importação do GitHub desses dados, realiza algumas manipulações em sua estrutura e gera quatro arquivos .csv  de maneira a indicar, respectivamente, casos e mortes por COVID-19 na Espanha, acumulados e diários, a nível de estado.

# Casos na Espanha

```python
url = 'https://raw.githubusercontent.com/datadista/datasets/master/COVID%2019/ccaa_covid19_casos.csv'
data_spain = pd.read_csv(url, error_bad_lines=False)
data_spain.drop(data_spain.columns[[0]],axis=1,inplace=True)
data_spain
```

```python
data_spain = data_spain.transpose()
data_spain = data_spain.rename(columns=data_spain.iloc[0]).drop(data_spain.index[0])
data_spain_daily = data_spain.diff()
data_spain_daily
```

Geração de dois arquivos .csv

```python
data_spain.to_csv('covid19-spain-region-cases-acc.csv')
data_spain_daily.to_csv('covid19-spain-region-cases-daily.csv')
```

# Mortes na Espanha

```python
url2 = 'https://raw.githubusercontent.com/datadista/datasets/master/COVID%2019/ccaa_covid19_fallecidos.csv'
data_spain_deaths = pd.read_csv(url2, error_bad_lines=False)
data_spain_deaths.drop(data_spain_deaths.columns[[0]],axis=1,inplace=True)
data_spain_deaths
```

```python
data_spain_deaths = data_spain_deaths.transpose()
data_spain_deaths = data_spain_deaths.rename(columns=data_spain_deaths.iloc[0]).drop(data_spain_deaths.index[0])
data_spain_deaths_daily = data_spain_deaths.diff()
data_spain_deaths_daily
```

Geração de dois arquivos .csv

```python
data_spain_deaths.to_csv('covid19-spain-region-deaths-acc.csv')
data_spain_deaths_daily.to_csv('covid19-spain-region-deaths-daily.csv')
```
