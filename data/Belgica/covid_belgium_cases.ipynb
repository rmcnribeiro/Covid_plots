```python
import pandas as pd
import numpy as np
```

```python
data_belgium = pd.read_excel('COVID19BE_20200402.xlsx')
data_belgium = data_belgium[data_belgium['PROVINCE'].notna()]
data_belgium = data_belgium[data_belgium['DATE'].notna()]
data_belgium
```

```python
covid_belgium = pd.DataFrame(np.zeros((48,11)))
covid_belgium.columns = data_belgium['PROVINCE'].unique()
covid_belgium.index = data_belgium['DATE'].unique()
```

```python
data_belgium = data_belgium.groupby(['DATE', 'PROVINCE']).sum().reset_index()
data_belgium = data_belgium.set_index('DATE')
data_belgium
```

```python
for index, row in data_belgium.iterrows():
    covid_belgium.at[index,row['PROVINCE']] = row['CASES']
```

```python
covid_belgium
```

```python
covid_belgium = covid_belgium[:-1]
covid_belgium.to_csv('covid19_belgium_cases.csv')
```
